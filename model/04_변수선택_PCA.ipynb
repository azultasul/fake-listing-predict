# ğŸ“ğŸ“ train + test

```python
import pandas as pd
import numpy as np

train = pd.read_csv('/Users/pycode/project/á„ƒá…¦á„‹á…µá„á…¥/train.csv')
train.info()
train.head()
# train.isna().sum()
```

```python

test = pd.read_csv('/Users/pycode/project/á„ƒá…¦á„‹á…µá„á…¥/test.csv')
test.info()
test.head()
test.isna().sum()
```

## ğŸ“ ë°ì´í„° íƒ€ì… ì§€ì •
* train, test í•©ì¹˜ê¸° ì „ì— ë°ì´í„° íƒ€ì… ì§€ì •   
train -> ë°ì´í„°íƒ€ì… : 'train'    
test -> ë°ì´í„°íƒ€ì… : 'test' 
* concatìœ¼ë¡œ train + test

```python
train['ë°ì´í„°íƒ€ì…'] = 'train'
test['ë°ì´í„°íƒ€ì…'] = 'test'
total_data = pd.concat([train.drop(columns='í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€'), test])

total_data.info()
```

## ğŸ“ train + test í›„ íŒŒìƒë³€ìˆ˜ ìƒì„±
* ë¯¸ê¸°ì¬
* ë°©ì¹˜ì¼ìˆ˜
* ì§‘ì„¸

```python
import datetime

# ë¯¸ê¸°ì¬
total_data['ë¯¸ê¸°ì¬'] = total_data.isna().sum(axis=1)

# ë°©ì¹˜ì¼ìˆ˜
last = datetime.datetime.strptime(total_data['ê²Œì¬ì¼'].max(), "%Y-%m-%d").date()
today = last + datetime.timedelta(days=1)

def get_delta_day(date):
    delta = today - datetime.datetime.strptime(date, "%Y-%m-%d").date()
    return delta.days

total_data['ë°©ì¹˜ì¼ìˆ˜'] = total_data['ê²Œì¬ì¼'].apply(lambda x: get_delta_day(x))

# ì§‘ì„¸
total_data['ì§‘ì„¸'] = total_data['ê´€ë¦¬ë¹„'] * 10000 + total_data['ì›”ì„¸']

total_data.head()
```

## ğŸ“ train + test í›„ ê²°ì¸¡ì¹˜ ì²˜ë¦¬

### 1. ë°©ìˆ˜, ìš•ì‹¤ìˆ˜   
  -> ê°€ì¥ ë§ì€ ìˆ˜ì¸ 1ë¡œ ì±„ì›€

```python
room = total_data['ë°©ìˆ˜'].value_counts()
bath = total_data['ìš•ì‹¤ìˆ˜'].value_counts()
print(room)
print(bath)

total_data[['ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜']] = total_data[['ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜']].fillna(1.0)
```

### 2. ì „ìš©ë©´ì    
  -> ë°©ìˆ˜, ìš•ì‹¤ìˆ˜ ì¡°í•©ìœ¼ë¡œ ì „ìš©ë©´ì  í‰ê·  ë° ì ìš©

```python
dropna_total = total_data.dropna(subset=['ì „ìš©ë©´ì '], how='any', axis=0)
new_data = pd.DataFrame(dropna_total, columns=('ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜', 'ì „ìš©ë©´ì '))

mean_data = new_data.groupby(['ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜'], as_index=False).mean()
print(mean_data)

for index, row in total_data.iterrows():
    if (pd.isna(row['ì „ìš©ë©´ì '])):
        filterTrain = mean_data[(mean_data['ë°©ìˆ˜'] == row['ë°©ìˆ˜']) & (mean_data['ìš•ì‹¤ìˆ˜'] == row['ìš•ì‹¤ìˆ˜'])]
        total_data.at[index, 'ì „ìš©ë©´ì '] = filterTrain['ì „ìš©ë©´ì '].values[0]

total_data.head()
```

### 3. í•´ë‹¹ì¸µ, ì´ì¸µ   
  -> í‰ê· ì„ ë°˜ì˜¬ë¦¼í•˜ì—¬ ì ìš©

```python
room = round(total_data['í•´ë‹¹ì¸µ'].mean())
bath = round(total_data['ì´ì¸µ'].mean())
print(room)
print(bath)
total_data['í•´ë‹¹ì¸µ'] = total_data['í•´ë‹¹ì¸µ'].fillna(room)
total_data['ì´ì¸µ'] = total_data['ì´ì¸µ'].fillna(bath)

total_data.isna().sum()
```

### 4. ì´ì£¼ì°¨ëŒ€ìˆ˜
-> í‰ê· ì„ ë°˜ì˜¬ë¦¼í•˜ì—¬ ì ìš©

```python
parking = round(total_data['ì´ì£¼ì°¨ëŒ€ìˆ˜'].mean())
total_data['ì´ì£¼ì°¨ëŒ€ìˆ˜'] = total_data['ì´ì£¼ì°¨ëŒ€ìˆ˜'].fillna(parking)
```

## ğŸ“ ë³´ì¦ê¸ˆ, ì›”ì„¸, ê´€ë¦¬ë¹„ ë‹¨ìœ„ ì •ë¦¬

```python
total_data['ë³´ì¦ê¸ˆ'] = total_data['ë³´ì¦ê¸ˆ'] / 10000
total_data['ì›”ì„¸'] = total_data['ì›”ì„¸'] / 10000
total_data['ì§‘ì„¸'] = total_data['ì§‘ì„¸'] / 10000
# total_data['ê´€ë¦¬ë¹„'] = total_data['ê´€ë¦¬ë¹„'] * 10000
total_data.head()
```

# ğŸ“ğŸ“ Encoding
-> ë ˆì´ë¸” ì¸ì½”ë”© (Label encoding) ì‚¬ìš©

```python
from sklearn.preprocessing import LabelEncoder

def encode_features(dataDF):
    features = ['ë§¤ë¬¼í™•ì¸ë°©ì‹', 'ë°©í–¥', 'ì£¼ì°¨ê°€ëŠ¥ì—¬ë¶€', 'ì¤‘ê°œì‚¬ë¬´ì†Œ', 'ì œê³µí”Œë«í¼']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature])
    
    return dataDF

total_data_encoded = encode_features(total_data)
total_data_encoded.head()
```

# ğŸ“ğŸ“ Feature Scaling
-> MinMaxScaler ì‚¬ìš©

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScalerê°ì²´ ìƒì„±
scaler = MinMaxScaler()
# MinMaxScaler ë¡œ ë°ì´í„° ì…‹ ë³€í™˜. fit() ê³¼ transform() í˜¸ì¶œ.  
cols_for_scale = ['ë³´ì¦ê¸ˆ', 'ì›”ì„¸', 'ì „ìš©ë©´ì ', 'í•´ë‹¹ì¸µ', 'ì´ì¸µ', 'ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜', 'ì´ì£¼ì°¨ëŒ€ìˆ˜', 'ê´€ë¦¬ë¹„', 'ë¯¸ê¸°ì¬', 'ë°©ì¹˜ì¼ìˆ˜', 'ì§‘ì„¸']
scaler.fit(total_data_encoded[cols_for_scale])
total_data_encoded[cols_for_scale] = scaler.transform(total_data_encoded[cols_for_scale])

# transform()ì‹œ scale ë³€í™˜ëœ ë°ì´í„° ì…‹ì´ numpy ndarryë¡œ ë°˜í™˜ë˜ì–´ ì´ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
total_data_scaled = pd.DataFrame(data=total_data_encoded, columns=total_data_encoded.columns)
print('featureë“¤ì˜ ìµœì†Ÿê°’')
print(total_data_scaled.min())
print('\nfeatureë“¤ì˜ ìµœëŒ“ê°’')
print(total_data_scaled.max())
total_data_scaled.head()
```

# ğŸ“ğŸ“ PCA (ì£¼ì„±ë¶„ ë¶„ì„)

```python
from sklearn.decomposition import PCA

# total_data_pca = total_data_scaled.drop(columns=['ID', 'ê²Œì¬ì¼', 'ë°ì´í„°íƒ€ì…'])
# total_data_scaled.head()

pca = PCA(n_components=2)

cols_for_scale = ['ë³´ì¦ê¸ˆ', 'ì›”ì„¸', 'ì „ìš©ë©´ì ', 'í•´ë‹¹ì¸µ', 'ì´ì¸µ', 'ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜', 'ì´ì£¼ì°¨ëŒ€ìˆ˜', 'ê´€ë¦¬ë¹„', 'ë¯¸ê¸°ì¬', 'ë°©ì¹˜ì¼ìˆ˜', 'ì§‘ì„¸']
pca.fit(total_data_scaled[cols_for_scale])
data_pca = pca.transform(total_data_scaled[cols_for_scale])
# total_data_scaled.head()

total_data_df = pd.DataFrame(data_pca, columns=['PCA1', 'PCA2'])
total_data_df.head()
total_data_pca = pd.concat([total_data_scaled.reset_index(drop=True), total_data_df], axis=1)
total_data_pca.head()
```

# ğŸ“ğŸ“ train, test ë°ì´í„° ë¶„ë¦¬

```python
total_data_prepro = total_data_pca
train_data = total_data_prepro[total_data_prepro['ë°ì´í„°íƒ€ì…'] == 'train']
test_data = total_data_prepro[total_data_prepro['ë°ì´í„°íƒ€ì…'] == 'test']
# print(train_data.head())
# print("-" * 10)
# print(test_data.head())

# def set_false(data) :
#     tttt = train[train['ID'] == data]
#     return tttt['í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€'].values[0]

# train_data["í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€"] = train_data["ID"].apply(lambda x : set_false(x))

train_data["í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€"] = train.set_index("ID").loc[train_data["ID"], "í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€"].values

# train_data = train_data.drop(columns=['ê²Œì¬ì¼', 'ë°ì´í„°íƒ€ì…'])
# test_data = test_data.drop(columns=['ê²Œì¬ì¼', 'ë°ì´í„°íƒ€ì…'])
train_data_mo = train_data.drop(columns=['ID', 'ê²Œì¬ì¼', 'ë°ì´í„°íƒ€ì…'])
test_data_mo = test_data.drop(columns=['ID', 'ê²Œì¬ì¼', 'ë°ì´í„°íƒ€ì…'])
print(train_data_mo.head())
print('-'*20)
print(test_data_mo.head())

# train_data, test_data íŒŒì¼ë¡œ ì €ì¥
# modified_train_data = train_data.drop(columns='ë°ì´í„°íƒ€ì…')
# modified_test_data = test_data.drop(columns='ë°ì´í„°íƒ€ì…')
# modified_train_data.to_csv("/Users/pycode/project/á„ƒá…¦á„‹á…µá„á…¥/modified_train.csv", index = False)
# modified_test_data.to_csv("/Users/pycode/project/á„ƒá…¦á„‹á…µá„á…¥/modified_test.csv", index = False)
```

# ğŸ“ğŸ“ XGBoost

```python
from xgboost import plot_importance
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline 

from matplotlib import rc
rc('font', family='AppleGothic')

from xgboost import XGBClassifier
```

```python
X_train = train_data_mo.iloc[:, :-1]
y_train = train_data_mo.iloc[:, -1]
X_test = test_data_mo.iloc[:, :]

X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=156)

print(X_tr.shape , X_val.shape)
```

## ğŸ“ ì‚¬ì´í‚·ëŸ° ë˜í¼

```python
xgb_wrapper = XGBClassifier(n_estimators=400, learning_rate=0.05, max_depth=3)
evals = [(X_tr, y_tr), (X_val, y_val)]
xgb_wrapper.fit(X_tr, y_tr, early_stopping_rounds=50, eval_metric='logloss', eval_set=evals, verbose=True)

ws50_preds = xgb_wrapper.predict(X_test)
ws50_preds_proba = xgb_wrapper.predict_proba(X_test)[:, 1]
```

```python
# fig, ax = plt.subplots(figsize=(10, 12))

# plot_importance(xgb_wrapper, ax=ax)
plot_importance(xgb_wrapper)
plt.show()
```

```python

# len(ws50_preds)
# ws50_preds
# test_data['ID']

df = pd.concat([pd.Series(test_data['ID'], name="ID"), pd.DataFrame(ws50_preds, columns=["í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€"])], axis=1)
print(df)

# df.to_csv("/Users/pycode/project/output.csv", index = False)
```
