```python
import pandas as pd
import numpy as np

train = pd.read_csv('/Users/pycode/project/á„ƒá…¦á„‹á…µá„á…¥/train.csv')
train.info()
train.head()
# train.isna().sum()
```

```python

test = pd.read_csv('/Users/pycode/project/á„ƒá…¦á„‹á…µá„á…¥/test.csv')
test.info()
test.head()
test.isna().sum()
```

## ğŸ“ ë°ì´í„° íƒ€ì… ì§€ì •
* train, test í•©ì¹˜ê¸° ì „ì— ë°ì´í„° íƒ€ì… ì§€ì •   
train -> ë°ì´í„°íƒ€ì… : 'train'    
test -> ë°ì´í„°íƒ€ì… : 'test' 
* concatìœ¼ë¡œ train + test

```python
train['ë°ì´í„°íƒ€ì…'] = 'train'
test['ë°ì´í„°íƒ€ì…'] = 'test'
total_data = pd.concat([train.drop(columns='í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€'), test])

total_data.info()
```

## ğŸ“ train + test í›„ íŒŒìƒë³€ìˆ˜ ìƒì„±
* ë¯¸ê¸°ì¬
* ë°©ì¹˜ì¼ìˆ˜
* ì§‘ì„¸

```python
import datetime

# ë¯¸ê¸°ì¬
total_data['ë¯¸ê¸°ì¬'] = total_data.isna().sum(axis=1)

# ë°©ì¹˜ì¼ìˆ˜
last = datetime.datetime.strptime(total_data['ê²Œì¬ì¼'].max(), "%Y-%m-%d").date()
today = last + datetime.timedelta(days=1)

def get_delta_day(date):
    delta = today - datetime.datetime.strptime(date, "%Y-%m-%d").date()
    return delta.days

total_data['ë°©ì¹˜ì¼ìˆ˜'] = total_data['ê²Œì¬ì¼'].apply(lambda x: get_delta_day(x))

# ì§‘ì„¸
total_data['ì§‘ì„¸'] = total_data['ê´€ë¦¬ë¹„'] * 10000 + total_data['ì›”ì„¸']

total_data.head()
```

## ğŸ“ train + test í›„ ê²°ì¸¡ì¹˜ ì²˜ë¦¬

### 1. ë°©ìˆ˜, ìš•ì‹¤ìˆ˜   
  -> ê°€ì¥ ë§ì€ ìˆ˜ì¸ 1ë¡œ ì±„ì›€

```python
room = total_data['ë°©ìˆ˜'].value_counts()
bath = total_data['ìš•ì‹¤ìˆ˜'].value_counts()
print(room)
print(bath)

total_data[['ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜']] = total_data[['ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜']].fillna(1.0)
```

### 2. ì „ìš©ë©´ì    
  -> ë°©ìˆ˜, ìš•ì‹¤ìˆ˜ ì¡°í•©ìœ¼ë¡œ ì „ìš©ë©´ì  í‰ê·  ë° ì ìš©

```python
dropna_total = total_data.dropna(subset=['ì „ìš©ë©´ì '], how='any', axis=0)
new_data = pd.DataFrame(dropna_total, columns=('ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜', 'ì „ìš©ë©´ì '))

mean_data = new_data.groupby(['ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜'], as_index=False).mean()
print(mean_data)

for index, row in total_data.iterrows():
    if (pd.isna(row['ì „ìš©ë©´ì '])):
        filterTrain = mean_data[(mean_data['ë°©ìˆ˜'] == row['ë°©ìˆ˜']) & (mean_data['ìš•ì‹¤ìˆ˜'] == row['ìš•ì‹¤ìˆ˜'])]
        total_data.at[index, 'ì „ìš©ë©´ì '] = filterTrain['ì „ìš©ë©´ì '].values[0]

total_data.head()
```

### 3. í•´ë‹¹ì¸µ, ì´ì¸µ   
  -> í‰ê· ì„ ë°˜ì˜¬ë¦¼í•˜ì—¬ ì ìš©

```python
room = round(total_data['í•´ë‹¹ì¸µ'].mean())
bath = round(total_data['ì´ì¸µ'].mean())
print(room)
print(bath)
total_data['í•´ë‹¹ì¸µ'] = total_data['í•´ë‹¹ì¸µ'].fillna(room)
total_data['ì´ì¸µ'] = total_data['ì´ì¸µ'].fillna(bath)

total_data.isna().sum()
```

### 4. ì´ì£¼ì°¨ëŒ€ìˆ˜
-> í‰ê· ì„ ë°˜ì˜¬ë¦¼í•˜ì—¬ ì ìš©

```python
parking = round(total_data['ì´ì£¼ì°¨ëŒ€ìˆ˜'].mean())
total_data['ì´ì£¼ì°¨ëŒ€ìˆ˜'] = total_data['ì´ì£¼ì°¨ëŒ€ìˆ˜'].fillna(parking)
```

## ğŸ“ ë³´ì¦ê¸ˆ, ì›”ì„¸, ê´€ë¦¬ë¹„ ë‹¨ìœ„ ì •ë¦¬

```python
total_data['ë³´ì¦ê¸ˆ'] = total_data['ë³´ì¦ê¸ˆ'] / 10000
total_data['ì›”ì„¸'] = total_data['ì›”ì„¸'] / 10000
total_data['ì§‘ì„¸'] = total_data['ì§‘ì„¸'] / 10000
# total_data['ê´€ë¦¬ë¹„'] = total_data['ê´€ë¦¬ë¹„'] * 10000
total_data.head()
```

# ğŸ“ğŸ“ Encoding
-> ë ˆì´ë¸” ì¸ì½”ë”© (Label encoding) ì‚¬ìš©

```python
from sklearn.preprocessing import LabelEncoder

def encode_features(dataDF):
    features = ['ë§¤ë¬¼í™•ì¸ë°©ì‹', 'ë°©í–¥', 'ì£¼ì°¨ê°€ëŠ¥ì—¬ë¶€', 'ì¤‘ê°œì‚¬ë¬´ì†Œ', 'ì œê³µí”Œë«í¼']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature])
    
    return dataDF

total_data = encode_features(total_data)
total_data.head()
```

# ğŸ“ğŸ“ Feature Scaling
-> MinMaxScaler ì‚¬ìš©

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScalerê°ì²´ ìƒì„±
scaler = MinMaxScaler()
# MinMaxScaler ë¡œ ë°ì´í„° ì…‹ ë³€í™˜. fit() ê³¼ transform() í˜¸ì¶œ.  
cols_for_scale = ['ë³´ì¦ê¸ˆ', 'ì›”ì„¸', 'ì „ìš©ë©´ì ', 'í•´ë‹¹ì¸µ', 'ì´ì¸µ', 'ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜', 'ì´ì£¼ì°¨ëŒ€ìˆ˜', 'ê´€ë¦¬ë¹„', 'ë¯¸ê¸°ì¬', 'ë°©ì¹˜ì¼ìˆ˜', 'ì§‘ì„¸']
scaler.fit(total_data[cols_for_scale])
total_data[cols_for_scale] = scaler.transform(total_data[cols_for_scale])

# transform()ì‹œ scale ë³€í™˜ëœ ë°ì´í„° ì…‹ì´ numpy ndarryë¡œ ë°˜í™˜ë˜ì–´ ì´ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜
total_data_scaled = pd.DataFrame(data=total_data, columns=total_data.columns)
print('featureë“¤ì˜ ìµœì†Ÿê°’')
print(total_data_scaled.min())
print('\nfeatureë“¤ì˜ ìµœëŒ“ê°’')
print(total_data_scaled.max())
total_data_scaled.head()
```
